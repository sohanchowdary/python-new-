```python
#compound
p = float(input(""))
r = float(input(""))
n = int(input(""))
t = int(input(""))
amount = p * (1 + r/(100 * n))**(n*t)
ci = amount - p;
print("Compound interest = ",round(ci, 2))
```

```python
#fibonacci
n = int(input(""))
n1, n2 = 0, 1
count = 0
if n <= 0:
   print("Please enter a positive integer")
elif n == 1:
   print(n1)
else:
   print("Fibonacci sequence:")
   while count < n:
       print(n1)
       nth = n1 + n2
       n1 = n2
       n2 = nth
       count += 1
```

```python
#gcd
a=int(input("enter a number"))
b=int(input("enter a number"))
k=a if a<b else b
while True:
    if a%k==0 and b%k==0:
     break
    k -=1
print(k)
```

```python
#pallindrome
a = str(input(""))
a = a.casefold()
b = reversed(a)
if list(a) == list(b):
   print("The string is a palindrome.")
else:
   print("The string is not a palindrome.")
```

```python
#perfect
n=int(input("Enter the number: "))  
sum=0  
for i in range(1,n):  
    if (n%i==0):  
        sum=sum+i  
if(sum==n):  
    print("perfect number")  
else:  
    print("not a perfect number")  
```
