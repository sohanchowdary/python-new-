### **FIRST**



```python
#check if string starts with sohan ends with assignment
import re

txt = "sohan completed his assignment"
vb = re.search("^sohan.*assignment$", txt)
if vb:
  print("yes string is available")
else:
  print("no")
```

### **SECOND**

```python
#using findall function in regx module
import re

sohan = "sohan writing his assignment"
p = re.findall("writing", sohan)
print(p)
```

### **THIRD**

```python
#using search function in python
import re

txt = "Sohan gets an O grade in python"
x = re.search("Sohan", txt)
if x:
  print("there is a match")
else:
  print("there is no match")
```

### **fourth**

```python
#using split function in python
import re

txt = "Sohan gets an O grade in python"

x = re.split("\s", txt)
print(x)
```

### **fifth**

```python
#using sub function in python
import re

txt = "sohan gets an O grade in python"
x = re.sub("\s", "9", txt)
print(x)
```
